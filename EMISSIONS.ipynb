Toplotna vrijednost utrošenog uglja iznosila je ukupno 12 122 (kJ/kg)

Potrošnja uglja u ovim elektranama po izvještaju EP BiH je 5095700 tona.

```python
coal=5095700*1000#kg
hd=12122#kj/kg
coal_en=coal*hd/1000000
print("Potrošnja uglja u GJ",coal_en)
```

```python
NOX=12077*1000000/coal_en
print("NOX",NOX,"g/GJ")
SO2=120978*1000000/coal_en
print("SO2",SO2,"g/GJ")
pm=581*1000000/coal_en
print("pm",pm,"g/GJ")
co2=4807737*100000/coal_en
print("CO2",co2,"g/GJ")
```

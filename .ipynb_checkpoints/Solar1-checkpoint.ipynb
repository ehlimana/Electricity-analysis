U opticaju su ostale male i velike solarne elektrane (osjenčene zelenim u .xslx. fajlu registar-ed-novembar-2021.xlsx - list: "Produzenje_solar"), s tim da nije jasno zašto su neki redovi potpuno zeleni, dok su za pojedine investitore podaci neosjenčeni. *Napomena: S obzirom da su u Sunny portal bazi samo veće FNE sa područja FBiH, nije razmatran registar koji se odnosi na dozovole iz RS-a. U bazi postoje podaci za FNE na području opštine Pale, te Banjaluke i Prnjavora, ali su kapaciteti mali. # Analiza registara privilegovanih proizvođača - 2021. na navedenim lokacijama

Lista privilegovanih proizvođača je preuzeta sa stranice Operatora za OIEiEK http://operatoroieiek.ba/proizvodaci-el-energije-iz-oieiek/privilegovani-proizvodaci/.

Privilegovani proizvođač električne energije, u smislu Zakona o korištenju obnovljivih izvora energije i efikasne kogeneracije (”Službene novine Federacije BiH”, br. 70/13 i 5/14) (Zakon), je kvalifikovani proizvođač koji ima pravo da se od njega po garantovanim cijenama otkupe ukupne količine proizvedene električne energije tokom unaprijed utvrđenog perioda.

Status privilegovanog proizvođača može dobiti imalac Dozvole za rad – licence za obavljanje djelatnosti proizvodnje električne energije, kojem je, u skladu sa članom 20. stav (2) Zakona, utvrđen status kvalifikovanog proizvođača.

Zahtjev za dobijanje statusa privilegovanog proizvođača, kvalifikovani proizvođač podnosi Operatoru za OIEiEK uz odgovarajuće dokaze u skladu sa pozitivnim zakonskim propisima koji regulišu datu oblast.

Kvalifikovani proizvođač koji je, rješenjem Operatora za OIEiEK, stekao status privilegovanog proizvođača, ima pravo da sa Operatorom za OIEiEK zaključi ugovor o otkupu ukupnog iznosa ili dijela proizvedene električne energije u postrojenjima koja koriste obnovljive izvore energije i efikasnu kogeneraciju.

```python
registar_priv=open_the_file(r"FILES\SOLAR\Registar_privil_proizv_30.9.2021.xlsx",1,0)
```

```python
registar_priv.head(2) #prva 2 reda 
```

Format Registra privilegovaniih proizvođača je nešto drugačiji u odnosu na format Registra imaoca energetskih dozvola, te se trenutno ne mogu porediti po sadržaju jer sadržaj u tabelama nije ujednačen i trenutno nije moguće automatski izvršiti komparaciju. Zato će se registar privilegovanih proizvođača podijeliti na oblasti sa poznatim lokacijama koje mogu biti sadržane u Nazivu podnosioca zahtjeva ili Nazivu postrojenja jer često podnosioci zahtjeva nisu u području lokacije postrojenja. Za Stolac će se pretraživanje vršiti po lokacijama: Hodovo, Pješivac, Stolac.
Sa ovakvim načinon selekcije može se vidjeti da je broj 

```python
registar_priv_STOLAC=registar_priv[(registar_priv["Naziv podnosioca zahtjeva"].str.contains("Stolac",case=False))|(registar_priv["Naziv postrojenja"].str.contains("Hodovo|Pješivac|Stolac",case=False))]
registar_priv_STOLAC
```

```python
FNE=registar_priv_STOLAC[pd.to_datetime(registar_priv_STOLAC["Datum podnošenja\nzahtjeva"]).dt.year<2019]
FNE.loc[:,"Instalisana\nsnaga (kW)"] = FNE.loc[:,"Instalisana\nsnaga (kW)"].str.replace(",", ".").astype(float)
capacity_sum=FNE["Instalisana\nsnaga (kW)"].sum(axis=0)
print("Broj FNE sa kompletiranim zahtjevom prije 2019.:",len(FNE), "\tUkupno instalisana snaga:",capacity_sum,"kW")

FNE=registar_priv_STOLAC[pd.to_datetime(registar_priv_STOLAC["Datum podnošenja\nzahtjeva"]).dt.year<2020]
FNE.loc[:,"Instalisana\nsnaga (kW)"] = FNE.loc[:,"Instalisana\nsnaga (kW)"].str.replace(",", ".").astype(float)
capacity_sum=FNE["Instalisana\nsnaga (kW)"].sum(axis=0) 
print("Broj FNE sa kompletiranim zahtjevom prije 2020.:",len(FNE), "\tUkupno instalisana snaga:",capacity_sum,"kW")

FNE=registar_priv_STOLAC[pd.to_datetime(registar_priv_STOLAC["Datum podnošenja\nzahtjeva"]).dt.year<2021]
FNE.loc[:,"Instalisana\nsnaga (kW)"] = FNE.loc[:,"Instalisana\nsnaga (kW)"].str.replace(",", ".").astype(float)
capacity_sum=FNE["Instalisana\nsnaga (kW)"].sum(axis=0) 
print("Broj FNE sa kompletiranim zahtjevom prije 2021.:",len(FNE), "\tUkupno instalisana snaga:",capacity_sum,"kW")
```

## Priprema fajlova

Prije nego što se krene u analizu, fajlovi sa satnim podacima za 5 referentnih FNE će se preurediti, te će se svakom podatku u godini dati vremenski otisak u vidu datuma i sata. Također, neki podaci postoje na 15-minutnoj rezoluciji, te će se u tom slučaju uzeti samo oni podaci koji odgovaraju cijelom satu. 

```python
def drop_empty_rows(filex):
    filex.drop(filex[filex[0]=='Diagram values'].index,inplace=True)
    filex.dropna(subset=[0],inplace=True)
    return filex #brisanje praznih redova
```

```python
#dodavanje vremenskog otiska
def assign_timestamp(filex,start,end,freq):
    filex=filex.assign(Timestamp=pd.date_range(start,end, freq=freq).values)
    return filex
```

```python
def set_index(filex):
    filex=filex.set_index(filex['Timestamp'].astype('datetime64[ns]'))
    return filex
#index
```

```python
from datetime import datetime
def assign_date(filex):
    filex=filex.assign(Date=filex.index.to_pydatetime())
    return filex

#dodjela datuma
```

Na području Stoca do 2019. godine instalisana snaga iznosi 2681 kW, a snaga referentne FNE iznosi 150 kW. Prema tome, broj ekvivalentnih elektrana je (2681/150 = 17,87). Zatim se godišnja proizvodnja za referentnu FNE pomnoži sa ovim brojem.
Nedoumica je gdje uključiti nesvrsate FNE iz registar_OSTALO. One su zasad uključene u dio koji odgovara referentnoj FNE iz Stoca jer se na taj način dobije približna proizvodnja onoj koja se navodi u izvještaju FERK-a za 2019. (22 GWh) i 2020. godinu (33 GWh).

```python
Total_production_FBIH_2019=(file_total_2019["Krenica1PowerMean values  [kW]"].sum(axis=0)*3.4+file_total_2019["Hodovo1PowerMean values  [kW]"].sum(axis=0)*(18+41.436)+file_total_2019["Enelectronic - Power, kW"].sum(axis=0)*2.44+file_total_2019["SE Energy - Power, kW"].sum(axis=0)*2+file_total_2019["CP Energy - Power, kW"].sum(axis=0)*(1.83))/1000000
print("Ukupna proizvodnja FBiH u 2019.",Total_production_FBIH_2019,"GWh" )
```
